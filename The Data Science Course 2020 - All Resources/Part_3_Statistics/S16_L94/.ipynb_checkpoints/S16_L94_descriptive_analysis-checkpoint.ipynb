```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from math import ceil as roundup
from numpy import nan
from matplotlib.ticker import PercentFormatter
```

```python
# loading data
sales = pd.read_csv("dataCSV.csv").drop('Unnamed: 9', axis = 1)
```

```python
# changing the columns names
sales.columns = sales.columns.str.replace(' ', '_').str.lower().str.replace('\(ft.\)', 'ft').str.replace("_#", '')

# cleaning column sales
sales['price'] = sales.price.str.replace('$', "").str.replace(',', '').astype(float)

sales.info()
```

# Task 2: Create a frequency distribution graph representing the price (that is a histogram with the highest possible number of bins - 272). Use data on all apartments and offices, no matter if sold or not.

# Task 3: Create a histogram which represents the Price variable. Choose interval width (bins) of length $100,000. If you don't know how to do that, refer to the Course notes on descriptive statistics provided with the first lecture in this section.

```python
# separating in group of 100,000
minimum = sales.price.min()
maximum = sales.price.max()
bins = roundup((maximum - minimum)/100000)

plt.figure(figsize = (16, 6))
sns.distplot(sales.price, bins = 5, kde = False)
sns.despine()
```

# Task 4: Interpret the results.

# Task 5: Create a scatter plot showing the relationship between Price and Area. Use the data on all properties, no matter if sold or not. Interpret the results.

```python
plt.figure(figsize = (16, 6))
ax = sns.scatterplot(x = sales.area_ft, y = sales.price, hue = sales.state)
ax.set_xlabel('area(ft²)')
ax.set_ylabel('price (dollars)')
ax.set_title('price in function of area')
sns.despine()
plt.show()
```

```python
# correlation between area in ft² and the price of the house
sales[['area_ft', 'price']].corr().loc['price', 'area_ft']
```

There is a high positive correlation between the size of the house and the price.

# Task 6: Create a frequency distribution table, where you list all the countries from which the company has buyers (country vs number of buyers). Count the absolute frequency, the relative frequency and the cumulative frequency. 

```python
plt.figure(figsize = (16, 5))
salesCountry = sales.country.value_counts(dropna = False).rename(index = {nan: 'Empty'})

ax = salesCountry.plot.bar()
sns.despine()

plt.show()
```

# Task 7: Create a Pareto diagram representing the data.

```python
total = salesCountry.sum()
Sum = 0
dcCountry = dict()

for country, value in salesCountry.iteritems():
    Sum += value
    dcCountry[country] = round(Sum/total*100, 2)
    
accPerc = pd.Series(dcCountry)
```

```python
fig, ax = plt.subplots(figsize = (16, 6))

ax.bar(salesCountry.index, salesCountry, label = 'Sales')
ax.set_ylabel('Total Sales')
ax2 = ax.twinx()
ax2.plot(accPerc.index, accPerc, color = 'r', marker = 'd', label = 'Sum Percentage')
ax2.set_ylabel('Pareto')
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylim(0)
fig.legend()
ax.set_xlabel('countries')
sns.despine()
ax.set_title('Sales by Country')

for country in accPerc.index:
    ax2.annotate(str(accPerc.loc[country]) + '%', xy=(country, accPerc.loc[country] - 4), color = 'black')

plt.show()
```

# Task 8: Calculate the mean, median, mode, skewness, variance and standard deviation of Price for all properties, no matter if sold or not.

```python
statiscSold = dict()
statiscSold['mean'] = sales.price.mean()
statiscSold['median'] = sales.price.median()
statiscSold['mode'] = sales.price.mode()[0]
# normalized by N-1.
statiscSold['skewness'] = sales.price.skew()
# normalized by N-1 by default. This can be changed using the ddof argument
statiscSold['standard deviation'] = sales.price.std()

statiscSold
```

# Task 9: Interpret the measures.

The median and the mean are not too far away that represent not a lot of outliers, the skewness higher than one represent a right tail on the data. The mode means very little in this case because that there is only one value that repeat and the standard deviation represents less tem 33% of the mean that represent that the data is not very spreed.

# Task 10: Calculate the covariance and correlation coefficient between Price and Area, no matter if the property is sold or not. Is the result in line with the scatter plot?

```python
# correlation
sales[['area_ft', 'price']].corr().loc['price', 'area_ft']
```

There is a high positive correlation between the size of the house and the price. That is what the scatter plot represents.
